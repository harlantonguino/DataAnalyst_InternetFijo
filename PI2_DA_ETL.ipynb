# 9. Acceso por Velocidad por Localizacion

# Dataset Nacion
### Ingresos + Penetracion + Banda + Acceso por Tecnologia + Promedio Velocidad + Acceso por Velocidad

```python
dfNacion = dfIngresos.merge(dfPenetracionTotales, how='inner', on=['Año', 'Trimestre'], suffixes=('_a', '_b'))
dfNacion = dfNacion.merge(dfBAFDialTotales, how='inner', on=['Año', 'Trimestre'], suffixes=('_a', '_b'))
dfNacion = dfNacion.merge(dfAccesosTecnologiaTotales, how='inner', on=['Año', 'Trimestre'], suffixes=('_a', '_b'))
dfNacion = dfNacion.merge(dfHistoricoVelocidadTotales, how='inner', on=['Año', 'Trimestre'], suffixes=('_a', '_b'))
dfNacion = dfNacion.merge(dfAccesosVelocidadTotales, how='inner', on=['Año', 'Trimestre'], suffixes=('_a', '_b'))
```
