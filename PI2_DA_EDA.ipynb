# Valores porcentuales respecto al total de accesos

Se calcula valores porcentuales respecto al total de los accesos. dividido en tres secciones (Banda ancha fija %,	Dial up %), (ADSL %,	Cablemodem %,	Fibra óptica %,	Wireless %,	OtrosTecno %), (Hasta 512 kbps %,	Entre 512 Kbps y 1 Mbps %,	Entre 1 Mbps y 6 Mbps %,	Entre 6 Mbps y 10 Mbps %,	Entre 10 Mbps y 20 Mbps %,	Entre 20 Mbps y 30 Mbps %,	Más de 30 Mbps %,	OtrosVel %)

```python
dfNacionPorc = dfNacionB.copy()
```

```python
# calcular valor porcentual respecto al total de accesos
dfNacionPorc['Banda ancha fija %'] = round(dfNacionPorc['Banda ancha fija'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Dial up %'] = round(dfNacionPorc['Dial up'] / dfNacionPorc['Total'] * 100,1)

dfNacionPorc['ADSL %'] = round(dfNacionPorc['ADSL'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Cablemodem %'] = round(dfNacionPorc['Cablemodem'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Fibra óptica %'] = round(dfNacionPorc['Fibra óptica'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Wireless %'] = round(dfNacionPorc['Wireless'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['OtrosTecno %'] = round(dfNacionPorc['OtrosTecno'] / dfNacionPorc['Total'] * 100,1)

dfNacionPorc['Hasta 512 kbps %'] = round(dfNacionPorc['Hasta 512 kbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Entre 512 Kbps y 1 Mbps %'] = round(dfNacionPorc['Entre 512 Kbps y 1 Mbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Entre 1 Mbps y 6 Mbps %'] = round(dfNacionPorc['Entre 1 Mbps y 6 Mbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Entre 6 Mbps y 10 Mbps %'] = round(dfNacionPorc['Entre 6 Mbps y 10 Mbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Entre 10 Mbps y 20 Mbps %'] = round(dfNacionPorc['Entre 10 Mbps y 20 Mbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Entre 20 Mbps y 30 Mbps %'] = round(dfNacionPorc['Entre 20 Mbps y 30 Mbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['Más de 30 Mbps %'] = round(dfNacionPorc['Más de 30 Mbps'] / dfNacionPorc['Total'] * 100,1)
dfNacionPorc['OtrosVel %'] = round(dfNacionPorc['OtrosVel'] / dfNacionPorc['Total'] * 100,1)
```

```python
# eliminar columnas
dfNacionPorc.drop(['Ingresos (miles de pesos)',	'Accesos por cada 100 hogares',	'Mbps (Media de bajada)',	'Total', 'ADSL',	'Cablemodem',	'Fibra óptica',	'Wireless',	'OtrosTecno', 'Banda ancha fija',	'Dial up',	'Hasta 512 kbps',	'Entre 512 Kbps y 1 Mbps',	'Entre 1 Mbps y 6 Mbps',	'Entre 6 Mbps y 10 Mbps',	'Entre 10 Mbps y 20 Mbps',	'Entre 20 Mbps y 30 Mbps',	'Más de 30 Mbps',	'OtrosVel'], axis=1, inplace=True)
```
